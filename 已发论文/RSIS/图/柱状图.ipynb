```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rcParams

config = {
            "font.family": 'serif',
            "font.size": 12,# 相当于小四大小
            "mathtext.fontset": 'stix',#matplotlib渲染数学字体时使用的字体，和Times New Roman差别不大
            "font.serif": ['SimSun'],#宋体
            'axes.unicode_minus': False # 处理负号，即-号
         }
rcParams.update(config)
```

```python
df = pd.read_excel('数据_.xlsx')
df
```

```python
df.columns[1:]
```

```python
df.iloc[0,0] = 'ASISO'
df
```

```python
plt.figure(figsize=(20,10),dpi=200)
# x = ['D1','D2','D3','D4','D5','D6']
size = 6
# x轴坐标, size=6, 返回[0, 1, 2, 3, 4, 5]
x = np.arange(size)
# 有5种类型的数据，n设置为5
total_width, n = 0.6, 5
# 每种类型的柱状图宽度
width = total_width / n
# 重新设置x轴的坐标
x = x - (total_width - width)/2


plt.bar(x,df.iloc[0,1:],width=width,label=df.iloc[0,0],color='royalblue')
plt.bar(x+width,df.iloc[1,1:],width=width,label=df.iloc[2,0],color='darkorange')
plt.bar(x+2*width,df.iloc[2,1:],width=width,label=df.iloc[3,0],color='darkgreen')
plt.bar(x+3*width,df.iloc[3,1:],width=width,label=df.iloc[4,0],color='grey')
# plt.bar(x+4*width,df.iloc[4,1:],width=width,label=df.iloc[5,0],color='darkred')
plt.legend(fontsize=15)
plt.xticks([0,1,2,3,4,5], df.columns[1:])
plt.grid(axis='y')

# 设置坐标轴标签
# y轴标签
plt.ylabel('MSE',
           labelpad=-17,                                         # 调整y轴标签与y轴距离
           y=0.98,                                             # 调整y轴标签的上下位置
           rotation=0,                                         # 调整位置为水平
           fontsize=15)                                        # 调整标签文字大小
plt.xticks(fontsize=15)
plt.yticks(fontsize=13)
```
