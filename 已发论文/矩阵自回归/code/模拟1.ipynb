```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 生成示例数据
np.random.seed(0)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 2, 100)
data3 = np.random.normal(-1, 0.5, 100)
data4 = np.random.normal(2, 1.5, 100)

# 创建多个变量DataFrame
df_multi = pd.DataFrame({
    'Group 1': data1,
    'Group 2': data2,
    'Group 3': data3,
    'Group 4': data4
})

# 将DataFrame转换为长格式（适合Seaborn的输入格式）
df_melted = df_multi.melt(var_name='Group', value_name='Value')

# 设置Matplotlib的样式
plt.style.use('seaborn-whitegrid')

# 绘制多个变量箱线图
plt.figure(figsize=(21, 18))
plt.subplot(3,3,1)
sns.boxplot(x='Group', y='Value', data=df_melted, palette='gray', fliersize=5, linewidth=1.5)
plt.title('Box Plot for Multiple Variables', fontsize=14)

plt.subplot(3,3,2)
sns.boxplot(x='Group', y='Value', data=df_melted, palette='gray', fliersize=5, linewidth=1.5)
plt.title('Box Plot for Multiple Variables', fontsize=14)

```

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成单变量数据
data = np.random.normal(0, 1, 100)

# 绘制单变量箱线图
plt.figure(figsize=(10, 6))
plt.boxplot(data)
plt.title('Box Plot for Single Variable')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.show()

# 生成多个变量数据
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 2, 100)
data3 = np.random.normal(-1, 0.5, 100)
data = [data1, data2, data3]

# 绘制多个变量箱线图
plt.figure(figsize=(10, 6))
plt.boxplot(data)
plt.title('Box Plot for Multiple Variables')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])
plt.show()
```
